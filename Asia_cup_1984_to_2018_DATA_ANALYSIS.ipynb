```python
##COMPARING FIRST INNING AND SECOND INNING AVERAGE RUN RATE

sns.set(rc={'figure.figsize':(15,5)})
sns.set(font_scale=1.5)
sns.lineplot(x="Match_id",y="fi_run_rate",data=df)
sns.lineplot(x="Match_id",y="si_run_rate",data=df)
```

```python
#Here we are showing how many teams are there in Team_1 of each country using bar plot
a=df["Team1"].value_counts()
print(a)
a.plot(kind="bar",color={"red","blue","yellow","green","violet","lightblue"})
```

```python
#Here we are showing how many teams are there in Team_2 of each country using bar plot
a=df["Team2"].value_counts()
print(a)
a.plot(kind="bar",color={"red","blue","yellow","green","violet","lightblue"})
```

```python
#Pie chart to show the percentage selection of batting and bowling
a=df.groupby("Selection").count()["Toss"]
print(a)
a.plot(kind="pie",subplots=True,autopct='%1.1f%%')
```

```python
#Plot to show the same team won toss and chosen batting and win the match
temp=df[df['Selection']=="Batting"]
a=temp[temp['Toss']==temp['Result']]
print(a)
a.groupby("Toss").count()["Result"].plot(kind='pie',subplots=True,autopct='%1.1f%%')
```

```python
#â€¢	Plot to show the same team won toss and chosen bowling and win the match
temp=df[df['Selection']=="Bowling"]
a=temp[temp['Toss']==temp['Result']]
print(a)
a.groupby("Toss").count()["Result"].plot(kind='pie',subplots=True,autopct='%1.1f%%')
```

```python
#Plot to show the percentage of team result
a=df.groupby('Result').count()["Toss"]
a.plot(kind="pie",subplots=True,autopct='%1.1f%%',title="Result")
```

```python
#plot to show the no of sixes of each team
a=df.groupby("Team1")
c=a["fi_6s"].sum()
b=df.groupby("Team2")
d=b["si_6s"].sum()
e=c+d
e.dropna()
e.plot(title="No of sixes of each team")
```

```python
#plot to show the no of four of each team
a=df.groupby("Team1")
c=a["fi_4s"].sum()
b=df.groupby("Team2")
d=b["si_4s"].sum()
e=c+d
e.dropna()
e.plot(color="red",title="No of fours of each team")
```
